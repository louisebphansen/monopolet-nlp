```python
import pandas as pd
import os
```

Columns are:
- episode_filename filename of the transcript from the archive; E.g. 2022-02-26-loerdag-26-feb-2022-11032200008.json
- intro_dillema_segment_id_estimation : int, estimation on where the dilemma starts in the transcript. It is a segment id from the transcript json
- first_person_dilemma and third_person_dilemma : generated dilemmas with all the context from the first person  and the third person respectively based on the transcript
- acceptable_solutions 	not_acceptable_solutions : list of extracted solutions from the transcript, generated as well. Acceptable - the ones advised in favour of doing so; Not acceptable - advised against doing so. The model is really struggles with the negation part, so in Not acceptable you will encounter a mix of "Do not do this" and "Do this"
- original_dilemma str, dilemma that was originally provided in the description of the episode on the podcast website.
